```python
# Passo a passo
# Passo 1: Importar base de dados
# Passo 2: Visualizar base de dados
# Passo 3: Corrigir erros da base de dados


# !pip install pandas numpy openpyxl nbformat ipykernel plotly
# Passo a passo do projeto
# Passo 1: Importar a base de dados

import pandas as pd
tabela = pd.read_csv("redesocial_saudemental.csv")

# passo 2 visualizar a tabela de dados 
tabela = tabela.drop(columns="User_ID")
#colunas inuteis não ajudam só atrapalham
display(tabela)
```

```python
# passo 3 corrigir os erros da base de dados
display(tabela.info())
#valores vazios - excluir as linhas que tem valores vazios
tabela = tabela.dropna()

display(tabela.info())

```

```python
# passo 4 Analise inicial dos sentimentos
# quantas pessoas deram nota positiva e negativa nas analises
display(tabela["Happiness_Index(1-10)"].value_counts())

# e em percentual

display(tabela["Happiness_Index(1-10)"].value_counts(normalize=True))
        
```

```python
# passo 5 Analise dos Indices de Felicidade das Pessoas 
# graficos/dashboards

import plotly.express as px

#criar o grafico


for coluna in tabela.columns:
    grafico = px.histogram(tabela,x=coluna,color="Happiness_Index(1-10)")
    # exibir o grafico
    grafico.show()
```

```python
# comparando as colunas:  

tabela = tabela[tabela["Sleep_Quality(1-10)"] != "Happiness_Index"]
tabela = tabela[tabela["Social_Media_Platform"]<="Happineess_Index(1-10)"]
tabela = tabela[tabela["Days_Without_Social_Media"]<=5]

display(tabela["Happiness_Index(1-10)"].value_counts())
# em percentual
display(tabela["Happiness_Index(1-10)"].value_counts(normalize=True))
```
